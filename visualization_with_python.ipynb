```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('users.csv')
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Group the data by age and sum the total income
income_by_age = df.groupby('age')['total_income'].sum().reset_index()

# Sort by income in descending order
income_by_age = income_by_age.sort_values(by='total_income', ascending=False)

# Plotting
plt.figure(figsize=(10, 6))
sns.barplot(x='age', y='total_income', data=income_by_age, palette='viridis')
plt.title('Total Income by Age')
plt.xlabel('Age')
plt.ylabel('Total Income')
plt.xticks(rotation=45)
plt.tight_layout()

# Save the chart as an image for PowerPoint
plt.savefig('total_income_by_age.png')

# Display the chart
plt.show()
```

```python
# Pivot the data to have categories as columns and gender as rows
spending_by_gender = df.pivot_table(index='gender', columns='expense_category', values='amount', aggfunc='sum').fillna(0)

# Plotting
spending_by_gender.plot(kind='bar', stacked=True, figsize=(12, 8), cmap='coolwarm')
plt.title('Gender Distribution Across Spending Categories')
plt.xlabel('Gender')
plt.ylabel('Total Spending')
plt.xticks(rotation=0)
plt.legend(title='Expense Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()

# Save the chart as an image for PowerPoint
plt.savefig('gender_distribution_spending.png')

# Display the chart
plt.show()
```
